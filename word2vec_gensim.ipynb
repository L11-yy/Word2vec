```python
# 词向量
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec
```

```python
# 读入训练集文件
data = pd.read_csv('train.csv')
# 转字符串数组
corpus = data['comment'].values.astype(str)
# 分词，再重组为字符串数组
corpus = [jieba.lcut(corpus[index]
                    .replace("，", "")
                    .replace("!", "")
                    .replace("！", "")
                    .replace("。", "")
                    .replace("~", "")
                    .replace("；", "")
                    .replace("？", "")
                    .replace("?", "")
                    .replace("【", "")
                    .replace("】", "")
                    .replace("#", "")
                    ) for index in range(len(corpus))]

#  使用Skip-Gram训练Word2Vec模型（sg=1表示Skip-Gram）
model = Word2Vec(corpus, sg=1, vector_size=300, window=5, min_count=3, workers=4)
```

```python
# 模型显示
print('模型参数：', model, '\n')
```

```python
#  输出"环境"的词向量及其形状
env_vector = model.wv.__getitem__('环境')
print("'环境'的词向量形状:", env_vector.shape)
print("'环境'的词向量(前5维):", env_vector[:5], '\n')
```

```python
#  输出与"好吃"语义最接近的3个词
print('与"好吃"最接近的3个词:', model.wv.most_similar('好吃', topn=3), '\n')
```

```python
#  计算相似度
print('"好吃"和"美味"的相似度=', model.wv.similarity('好吃', '美味'))
print('"好吃"和"蟑螂"的相似度=', model.wv.similarity('好吃', '蟑螂'), '\n')
```

```python
#  执行向量运算"餐厅+聚会-安静=？"
result = model.wv.most_similar(positive=['餐厅', '聚会'], negative=['安静'], topn=1)
print('"餐厅+聚会-安静"最相关结果:', result[0], '\n')
```

```python
print('最匹配的词是:', model.wv.most_similar(positive=['点赞', '不错'], negative=['难吃']), '\n')
print('相似度为=', model.wv.similarity('推荐', '好吃'), '\n')
print("'地道'的词向量(前5维):", model.wv.__getitem__('地道')[:5])
```
